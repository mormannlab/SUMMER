```python
from pathlib import Path
from random import randint
from pynwb import NWBHDF5IO, validate

from tabulate import tabulate

from write_nwb import *

import matplotlib.pyplot as plt

from tqdm import tqdm

data_dir = Path("refractored_data/patient_data")
save_dir = Path("nwb_dataset")

import pynwb
print(pynwb.__version__)
```

```python
pat_subset = []

sub_id_by_patient = np.arange(1, len(pat_subset)+1, dtype=int)

for i, patient_id in enumerate(tqdm(pat_subset)):  

    sub_id = sub_id_by_patient[i]

    print(patient_id, sub_id)
    patNWB = PatientNWB(patient_id=patient_id, sub_id=sub_id, data_dir=data_dir, save_dir=save_dir)
    print("Populating subject info .. ")
    patNWB.populate_subject()
    print("Populating movie binning data ..")
    patNWB.populate_movie_binning_data()
    print("Populating movie annotation/indication functions ..")
    patNWB.populate_movie_annotations()
    print("Populating movie PTS info ..")
    patNWB.populate_movie_pts()
    print("Populating electrode groups ..")
    patNWB.populate_electrode_groups()
    print("Populating electrodes ..")
    patNWB.populate_micro_channels()
    print("Populating units ..")
    patNWB.populate_units()

    print("Populating aligned annotations ..")
    patNWB.populate_aligned_annotations()
    print("Populating base annotations ..")
    patNWB.populate_base_annotations()
    print("Populating cleaned watchlogs ..")
    patNWB.populate_clean_watchlogs()
    print("Populating raw watchlogs ..")
    patNWB.populate_raw_watchlogs()

    patNWB.save_nwbfile()
    print('saved!')
```
